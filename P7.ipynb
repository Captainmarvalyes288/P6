```python
import pandas as pd ;
import numpy as np ;
```

```python
df = pd.read_csv('Market_Basket.csv' , header = None )
df.head()
```

```python
df.shape
```

```python
data_clean = df.dropna( how = 'all' )
data_clean = data_clean.apply( lambda x : [ item for item in x if item != '' and  pd.notna(item) ] )
data_clean
```

```python
from apyori import apriori
```

```python
data = data_clean.values.tolist()
```

```python
rules = apriori(transactions = data , min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
results = list(rules)
```

```python
def inspect(results):
    lhs         =[tuple(result[2][0][0])[0] for result in results]
    rhs         =[tuple(result[2][0][1])[0] for result in results]
    supports    =[result[1] for result in results]
    confidences =[result[2][0][2] for result in results]
    lifts        =[result[2][0][3] for result in results]
    return list (zip(lhs, rhs, supports, confidences, lifts))


resultStore = pd.DataFrame(inspect(results), columns = ["Left hand side", "Right hand side", "Support", "Confidence", "Lift"])
```

```python
resultStore
```

```python
top_rules = resultStore.nlargest(n=10, columns="Lift")
print(top_rules)
```
